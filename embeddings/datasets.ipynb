```python
from datasets import load_dataset

ds = load_dataset("stanfordnlp/snli")
```

```python
# Filter each split separately
ds_train_clean = ds["train"].filter(lambda x: x["label"] != -1)
ds_valid_clean = ds["validation"].filter(lambda x: x["label"] != -1)
ds_test_clean = ds["test"].filter(lambda x: x["label"] != -1)
```

```python
# Before/after counts
print(f"Train before: {len(ds['train'])}, after: {len(ds_train_clean)}")
print(f"Validation before: {len(ds['validation'])}, after: {len(ds_valid_clean)}")
print(f"Test before: {len(ds['test'])}, after: {len(ds_test_clean)}")
```

```python
# Reusable function
def load_snli_clean(split="train"):
    ds = load_dataset("stanfordnlp/snli", split=split)
    return ds.filter(lambda x: x["label"] != -1)

# Test it
train_clean = load_snli_clean("train")
print(f"Loaded clean train: {len(train_clean)} examples")
```
